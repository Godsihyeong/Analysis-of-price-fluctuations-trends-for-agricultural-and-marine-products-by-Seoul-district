```python
data = market[market['품목 이름'] == input('품목 이름을 입력하세요.')][['전처리 이전(년도-월)', '품목 정보(개수)', '가격(원)']]
data = data[data['품목 정보(개수)'] == int(input('품목 정보(개수)를 입력하세요.'))]
```

```python
df = data.groupby('전처리 이전(년도-월)')['가격(원)'].mean()
df
```

```python
df.sort_index().index.values # 2020년 3월에 결측치 발생

df['2020-03'] = (df['2020-02'] + df['2020-04'])/2
```

```python
import numpy as np
import pandas as pd
import os
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
from statsmodels.tsa.seasonal import seasonal_decompose 
#from pmdarima import auto_arima                        
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = df.reset_index()
df.columns = ['Month', 'Price']
df.Month = pd.to_datetime(df.Month)

df = df.set_index("Month").sort_index()
df.index.freq = 'MS'

df.head()
```

```python
plt.figure(figsize=(18,9))
plt.plot(df.index, df["Price"], linestyle="-")
plt.title('Monthly Egg Price')
plt.xlabel('Dates')
plt.ylabel('Total Price')
plt.show()
```

```python
train_data = df[:len(df)-12]
test_data = df[len(df)-12:]
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(train_data)
plot_pacf(train_data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

diff_1=train_data.diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_data, order=(0, 2, 2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
fore = model_fit.forecast(steps=12)
print(fore)
```
